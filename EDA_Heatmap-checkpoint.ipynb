```python
def set_fig_size():    
    f = plt.figure()
    f.set_figwidth(16)
    f.set_figheight(8)
```

```python
# Candel Sticks

import matplotlib.dates as mpl_dates
from mpl_finance import candlestick_ohlc

plt.style.use('ggplot')

ohlc = df.loc[:, ['datetime', 'open', 'high', 'low', 'close']]
ohlc['datetime'] = ohlc['datetime'].apply(mpl_dates.date2num)
ohlc = ohlc.astype(float)


fig, ax = plt.subplots(figsize=(30,20))

candlestick_ohlc(ax, ohlc.values, width=2, colorup='green', colordown='red', alpha=1)

ax.set_xlabel('Date')
ax.set_ylabel('Price')
fig.suptitle('Daily Candlestick Chart of NIFTY50')

date_format = mpl_dates.DateFormatter('%d-%m-%Y')
ax.xaxis.set_major_formatter(date_format)
fig.autofmt_xdate()
plt.show()
```

```python
#Applying Simple moving average
df['SMA30'] = df['close'].rolling(30).mean()
df['SMA70'] = df['close'].rolling(70).mean()
df['SMA140'] = df['close'].rolling(140).mean()
df['SMA280'] = df['close'].rolling(280).mean()
```

```python
set_fig_size()
plt.plot(df['datetime'],df['close'],label='Close')
plt.plot(df['datetime'],df['SMA30'],label='SMA30')
plt.plot(df['datetime'],df['SMA70'],label='SMA70')
plt.plot(df['datetime'],df['SMA140'],label='SMA140')
plt.plot(df['datetime'],df['SMA280'],label='SMA280')

plt.xlabel('Date')
plt.grid()
plt.legend()
```

```python
# Applying Exponential Moving average to calculate over a period of 20,50,100 and 200 days
set_fig_size()
df['20_EMA'] = df['close'].ewm(span=20, adjust=False).mean()
df['50_EMA'] = df['close'].ewm(span=50, adjust=False).mean()
df['100_EMA'] = df['close'].ewm(span=100, adjust=False).mean()
df['200_EMA'] = df['close'].ewm(span=200, adjust=False).mean()


plt.plot(df['datetime'],df['close'],label='Close')
plt.plot(df['datetime'],df['20_EMA'],label='EMA_20')
plt.plot(df['datetime'],df['50_EMA'],label='EMA_50')
plt.plot(df['datetime'],df['100_EMA'],label='EMA_100')
plt.plot(df['datetime'],df['200_EMA'],label='EMA_200')
plt.xlabel('Date')
plt.grid()
plt.legend()
```

From the graph above we see that wherever the four points are intersecting each other we see huge fluctuating trends of stock prices after evry point of intersection.

# Monthly percentage return plot using Heatmap

```python

#df['Month'] = pd.DatetimeIndex(df['datetime']).month
df['Month'] = df['datetime'].dt.strftime('%m')
df.info()
```

```python
df['year'] = pd.DatetimeIndex(df['datetime']).year
years = df['year'].unique()
years
```

```python
for i in range(len(df)):
  if df['Month'][i] == '01':
    df['Month'][i] = 'Jan'
  elif df['Month'][i] == '02':
    df['Month'][i] = 'Feb'
  elif df['Month'][i] == '03':
    df['Month'][i] = 'Mar'
  elif df['Month'][i] == '04':
    df['Month'][i] = 'Apr'
  elif df['Month'][i] == '05':
    df['Month'][i] = 'May'
  elif df['Month'][i] == '06':
    df['Month'][i] = 'Jun'
  elif df['Month'][i] == '07':
    df['Month'][i] = 'July'
  elif df['Month'][i] == '08':
    df['Month'][i] = 'Aug'
  elif df['Month'][i] == '09':
    df['Month'][i] = 'Sept'
  elif df['Month'][i] == '10':
    df['Month'][i] = 'Oct'
  elif df['Month'][i] == '11':
    df['Month'][i] = 'Nov'
  else:
    df['Month'][i] = 'Dec'
  
```

```python
df.head(30)
```

```python
df['Day'] = df['datetime'].dt.strftime('%d')
```

```python
df.head()
```

```python
year = df.year.unique()

offset = year[0]
```

```python
import numpy as np
open_pr, close_pr = np.zeros((len(year), 12)), np.zeros((len(year), 12))
```

```python
res = set()
for row in df.iterrows():
  #print(row[1])
  year, month = row[1].datetime.year, row[1].datetime.month
  if (year, month) not in res:
    res.add((year,month))
    open_pr[year - offset, month - 1] = row[1].open
  else:
    close_pr[year - offset, month - 1] = row[1].close
```

```python
open_pr, close_pr
```

```python
grid = (close_pr - open_pr) / open_pr
```

```python
months = ("Jan","Feb","Mar","Apr","May","June","July","Aug","Sept","Oct","Nov","Dec")
```

```python
heatmap = pd.DataFrame(grid, index = years, columns = list(months))
heatmap
```

```python

plt.figure(figsize=(28,12))
sns.heatmap(heatmap, annot=True, fmt='.2f',cmap="RdYlGn")
plt.xlabel("Month",{"fontsize":16})
plt.ylabel("Year",{"fontsize":16})
plt.title("Monthly Returns %", {"fontsize":18})
plt.show()
```
